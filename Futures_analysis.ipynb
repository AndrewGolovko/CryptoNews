```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('BTC-USD.csv')
df_cme = pd.read_csv('CME.csv')[:-2]
df_cboe = pd.read_csv('CBOE.csv')[:-1]
```

```python
df_cme.T
```

```python
xlim = (2750, df.shape[0])
step = 7
x = 'Date'
# y = ['High', 'Low', 'Open','Close']
y = ['Close']
df.plot(x, y, xlim=xlim, 
        ylim=(df[y].values[xlim[0]:xlim[1]].min(), df[y].values[xlim[0]:xlim[1]].max()), 
        figsize=(17,6))
# plt.grid()
plt.xticks(range(xlim[0], xlim[1], step), df.Date[xlim[0]:xlim[1]:step], rotation=60)

for cboe in df_cboe.Date:
    plt.plot(list(df.index[(cboe == df.Date)]) * 2,
             [df[y].values[xlim[0]:xlim[1]].min(), df[y].values[xlim[0]:xlim[1]].max()],
             'g:')
for cme in df_cme.Date:
    plt.plot(list(df.index[(cme == df.Date)]) * 2,
             [df[y].values[xlim[0]:xlim[1]].min(), df[y].values[xlim[0]:xlim[1]].max()],
             'b:')
plt.legend(y+['CBOE', 'CME']);
```
